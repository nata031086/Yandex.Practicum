# Декомпозиция

**Заказчик:** отдел маркетинга банка «Метанпромбанк»

Цель проекта: Анализ клиентов регионального банка и выделие сегментов клиентов, которые склонны уходить из банка для дальнейших мер по снижению оттока и сохранению текущих клиентов.

Входные данные: Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск. Данные о клиентах банка «Метанпром» содержатся в датасете /datasets/bank_scrooge.csv, доступ к которому можно получить из интернета по адресу https://code.s3.yandex.net/datasets/.\ Итак, в датасете представлены следующие колонки:
• USERID — идентификатор пользователя,
• score — баллы кредитного скоринга,
• city — город,
• gender — пол,
• age — возраст,
• equity — количество баллов собственности
• balance — баланс на счёте,
• products — количество продуктов, которыми пользуется клиент,
• credit_card — есть ли кредитная карта,
• last_activity — активный клиент,
• EST_SALARY — оценочный доход клиента,
• сhurn — признак оттока.

**Инструкция по выполнению проекта:**

**Шаг 1. Загрузка данных**\
•	Загрузим данные из CSV-файла в переменную\
•	Выведем первые строки таблицы набора данных\
•	Изучим общую информацию о наборе данных\
•	Подведем промежуточные итоги

**Шаг 2. Предобработка данных**\
• Заменим названия столбцов на удобные для нас\
• Найдем явные и неявные дубликаты в данных. Обработаем их, если такие значения присутствуют и это возможно\
• Проверим пропуски и типы данных, Откорректируем, если нужно\
• Обработаем аномальные значения\
• Подведем промежуточные итоги

**Шаг. 3. Исследовательский анализ данных**\
•	Исследуем портреты типичного текущего клиента банка и того клиента, который склонен уйти из банка (средние и медианы)\
•	Построим визуализации для непрерывных и категориальных величин\
•	Проведем корреляционный анализ\
•	Промежуточные выводы\

**Шаг. 4. Проверка гипотез**\
•	Проверим две статистические гипотезы:

        Гипотеза 1. о различии дохода между теми клиентами, которые ушли и теми, которые остались\
         H_0: доходы клиентов, которые ушли и теми, которые остались, равны\
         H_1: доходы клиентов, которые ушли и теми, которые остались, различаются
         
         Гипотеза 2. какое именно количество банковских продуктов, которым владеет клиент, влияет на отток клиентов из Банка
         H_0: На отток клиента из Банка влияет факт владения n-кол-вом банковских продуктов.\
         H_1: На отток клиента из Банка не влияет факт владения n-кол-вом банковских продуктов.
         
•	Промежуточный итог

**Шаг. 5.  Сегментируем клиентов Банка**\
• Сформируем сегменты\
• Посчитаем ключевые показатели по ним\
• Определим три лучших сегмента

**Шаг. 6. Выводы и рекомендации для Заказчика**\
. подведем итоги по анализу типичного клиента регионального банка\
. по выделенным сегментам клиентов, которые склонны уходить из банка, дадим рекомендации по снижению оттока и сохранению текущих клиентов.

**Шаг. 7. Подготовка презентации**\
Подготовим презентацию исследования для отдела маркетинга, выбирая важные тезисы и наблюдения, которые могут помочь в решении возникшей ситуации.\
Для создания презентации будем использовать любой удобный инструмент, но отправить презентацию нужно обязательно в формате PDF, прикрепив ссылку на файл в основном проекте.
